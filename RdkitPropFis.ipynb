üë®‚Äçüî¨üî¨üíä RDKit is an open-source library that allows for the analysis and manipulation of molecules in medicinal chemistry. Its utility in the field of computational chemistry is invaluable, as it allows for the calculation of important molecular properties that can be used in drug discovery and in understanding the structure-activity relationship of molecules.

üìä With RDKit, we can calculate essential molecular properties such as LogP, molecular weight, number of hydrogen bond donors and acceptors, among others. We can also estimate the molecule's permeability and half-life, and evaluate its potential for metabolism and possible side effects. Additionally, molecular features such as conformation, size, polarity, electrical charge, hydrophobicity/hydrophilicity, and flexibility can be calculated.

üîé By using RDKit in our code, we can quickly analyze the properties of a molecule and better understand its potential as a drug or its ability to interact with other compounds. Furthermore, we can use this information to optimize the molecular structure and improve its biological activity.

```python
!pip install rdkit

from rdkit import Chem
from rdkit.Chem import Descriptors

def calculate_properties(smiles):
    mol = Chem.MolFromSmiles(smiles)
    
    # Check if molecule is chiral
    if Chem.FindMolChiralCenters(mol):
        print("The molecule is chiral")
    else:
        print("The molecule is not chiral")
    
    # Calculate lipophilicity (logP)
    logp = Descriptors.MolLogP(mol)
    print("LogP: ", logp)
    
    # Calculate molecular weight
    weight = Descriptors.MolWt(mol)
    print("Molecular weight: ", weight)
    
    # Calculate number of hydrogen bond donors and acceptors
    hbd = Descriptors.NumHDonors(mol)
    print("Number of hydrogen bond donors: ", hbd)
    
    hba = Descriptors.NumHAcceptors(mol)
    print("Number of hydrogen bond acceptors: ", hba)

# Pass a SMILES string to the function to calculate the properties
calculate_properties("CC(C)(C)OC(=O)NC1CCC(CC1)N(C)c2ccccc2")
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import Draw
import numpy as np
import matplotlib.pyplot as plt

# Crear la mol√©cula a partir de un SMILES
smiles = 'CC(C)(C)OC(=O)NC1CCC(CC1)N(C)c2ccccc2'
mol = Chem.MolFromSmiles(smiles)

# A√±adir hidr√≥genos expl√≠citos a la mol√©cula
mol = Chem.AddHs(mol)

# Generar conformaciones con ETKDG
num_conformers = 50
AllChem.EmbedMultipleConfs(mol, num_conformers, AllChem.ETKDG())

# Optimizar las conformaciones con MMFF94s
mmff_properties = AllChem.MMFFGetMoleculeProperties(mol, mmffVariant="MMFF94s")
min_energy = np.inf
min_energy_conf_id = None
for conf in mol.GetConformers():
    ff = AllChem.MMFFGetMoleculeForceField(mol, mmff_properties, confId=conf.GetId())
    ff.Initialize()
    ff.Minimize()
    energy = ff.CalcEnergy()
    if energy < min_energy:
        min_energy = energy
        min_energy_conf_id = conf.GetId()
    print(f"Conformer {conf.GetId()+1}: Energy = {energy:.2f} kcal/mol")

# Crear lista con todas las conformaciones
confs = [mol.GetConformer(i) for i in range(mol.GetNumConformers())]



```

```python
from rdkit import Chem
from rdkit.Chem import AllChem

# Crear la mol√©cula a partir de un SMILES
mol = Chem.MolFromSmiles('CCO')

# Generar 10 conformaciones y optimizar su energ√≠a
num_conformers = 10
params = AllChem.ETKDGv3()
params.numThreads = -1
AllChem.EmbedMultipleConfs(mol, numConfs=num_conformers, params=params)
for conf in mol.GetConformers():
    AllChem.MMFFOptimizeMoleculeConfs(mol, maxIters=200, mmffVariant='MMFF94s')

# Calcular la energ√≠a de cada conformaci√≥n
energias = []
for idx in range(num_conformers):
    conf = mol.GetConformer(idx)
    ff = AllChem.MMFFGetMoleculeForceField(mol, pyMMFFMolProperties=AllChem.MMFFGetMoleculeProperties(mol), confId=idx)
    energia = ff.CalcEnergy()
    energias.append(energia)

# Imprimir los valores
print("Energ√≠as conformacionales:", energias)




```

```python
import matplotlib.pyplot as plt

# Conformers index for the x-axis
x = range(num_conformers)

# Create a bar chart of the conformer energies
plt.bar(x, energias)

# Add labels and title
plt.xticks(x, [f"Conf {i+1}" for i in x])
plt.ylabel("Energy (kJ/mol)")
plt.title("Conformational energy")

# Show the plot
plt.show()
```
